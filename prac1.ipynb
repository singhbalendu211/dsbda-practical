```python
import pandas as pd
```

```python
data={'Name':['Aryan','Rashmi','Yash','Kailash','Kasturi','Abhi','Shreya'],
      'Age':[17,18,17,17,19,19,17],
      'Gender':['M','F','M','M','F','M','F'],
      'Marks':[92,74,'NaN',88,'NaN',85,77]
     }

df=pd.DataFrame(data)
df
```

```python
cat=[]
con=[]

for i in df.columns:
    if(df[i].dtypes=='objects'):
        cat.append(i)
    else:
        con.append(i)
        
```

```python
print(cat)
```

```python
print(con)
```

```python
df
```

```python
c=avg=0
for ele in df['Marks']:
    if str(ele).isnumeric():
        c+=1
        avg+=ele
avg/=c

df=df.replace(to_replace='NaN',value=avg)
df    
```

```python
df.isna().sum().plot(kind='bar')
```

```python
df['Gender']=df['Gender'].map({'M':0, 'F':1}).astype(float)
```

```python
df
```

```python
df=df[df['Marks']>=75]
```

```python
df=df.drop(['Age'], axis=1)
```
