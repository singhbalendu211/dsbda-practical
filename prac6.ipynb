2. Compute Confusion matrix to find TP, FP, TN, FN, Accuracy, Error rate, Precision, Recall 
on the given dataset.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset
data = pd.read_csv('iris.csv')
```

```python
data
```

```python
# Prepare features (X) and target (y)
X = data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y = data['variety']
```

```python
# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Split the data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
# Initialize and train the Gaussian Naïve Bayes classifier
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = classifier.predict(X_test)
```

```python
# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
```

```python
# For multi-class, we'll calculate precision and recall for each class
precision_setosa = report['Setosa']['precision']
recall_setosa = report['Setosa']['recall']
precision_versicolor = report['Versicolor']['precision']
recall_versicolor = report['Versicolor']['recall']
precision_virginica = report['Virginica']['precision']
recall_virginica = report['Virginica']['recall']
```

```python
# Print results
print("Confusion Matrix:")
print(cm)
print("\nMetrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Error Rate: {error_rate:.4f}")
print("\nClass-wise Metrics:")
print(f"Setosa - Precision: {precision_setosa:.4f}, Recall: {recall_setosa:.4f}")
print(f"Versicolor - Precision: {precision_versicolor:.4f}, Recall: {recall_versicolor:.4f}")
print(f"Virginica - Precision: {precision_virginica:.4f}, Recall: {recall_virginica:.4f}")
```
