#  Data Visualization I
1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information 
about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to 
see if we can find any patterns in the data. 
2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger 
is distributed by plotting a histogram.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
# Display basic information
print(titanic.info())
print("\nFirst 5 rows:")
print(titanic.head())
```

```python
# Set style for plots
sns.set_style('whitegrid')
```

```python
# Create a figure with multiple plots
plt.figure(figsize=(15, 10))
```

```python
# Plot 1: Survival rate by class
plt.subplot(2, 2, 1)
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
```

```python
# Plot 2: Survival rate by gender
plt.subplot(2, 2, 2)
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Gender')
```

```python
#Plot 3: Age distribution of survivors vs non-survivors
plt.subplot(2, 2, 3)
sns.histplot(data=titanic, x='age', hue='survived', bins=30, kde=True, multiple='stack')
plt.title('Age Distribution: Survivors vs Non-Survivors')
```

```python
# Plot 4: Survival rate by embarkation port
plt.subplot(2, 2, 4)
sns.pointplot(x='embarked', y='survived', hue='class', data=titanic)
plt.title('Survival Rate by Embarkation Port and Class')
```

```python
plt.tight_layout()
plt.show()
```

```python
#Ticket Fare Distribution Analysis

plt.figure(figsize=(12, 6))

# Plot histogram of ticket fares
sns.histplot(data=titanic, x='fare', bins=30, kde=True, color='royalblue')
```

```python
# Add title and labels
plt.title('Distribution of Ticket Fares on Titanic', fontsize=14)
plt.xlabel('Fare (in currency units)', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
```

```python
# Highlight the expensive tickets
max_fare = titanic['fare'].max()
plt.axvline(x=max_fare, color='red', linestyle='--', 
            label=f'Most expensive ticket: {max_fare}')
plt.legend()

plt.show()
```
