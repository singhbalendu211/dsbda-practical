```python
import pandas as pd
import statistics as stat
```

```python
df=pd.read_csv('iris.csv')
df = df.rename(columns={'variety': 'species'})
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.groupby('species')[['sepal.length','sepal.width','petal.length','petal.width']].mean()
```

```python
df.groupby('species')[['sepal.length','sepal.width','petal.length','petal.width']].median()
```

```python
df.groupby('species')[['sepal.length','sepal.width','petal.length','petal.width']].min()
```

```python
df.groupby('species')[['sepal.length','sepal.width','petal.length','petal.width']].max()
```

```python
df.groupby('species')[['sepal.length','sepal.width','petal.length','petal.width']].quantile()
```

```python
df.groupby('species')[['sepal.length','sepal.width','petal.length','petal.width']].std()
```
